```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Loading dataset.
honey = pd.read_csv('vHoneyNeonic_v03.csv')
```

```python
sns.set(style="ticks")
sns.regplot(honey.nAllNeonic, honey.yieldpercol)
```

```python
fig, ax = plt.subplots()
ax = sns.pointplot(honey.year, honey.yieldpercol)
fig.set_size_inches(16,4)
```

```python
fig, ax = plt.subplots()
ax = sns.barplot(honey.year, honey.yieldpercol, honey.StateName)
fig.set_size_inches(16,4)
```

```python
sns.boxplot(honey.Region,honey.yieldpercol)
```
